***Write a Python program to convert temperature from Celsius to Fahrenheit using the formula: F = (C × 9/5) + 32.***


```python
celsius = float(input("Enter the Celsius = "))
fahrenheit = (celsius * 9/5) + 32
print(fahrenheit)

```

***Write a Python program to convert temperature from Fahrenheit to Celsius using the formula: C = (F - 32) × 5/9.***
